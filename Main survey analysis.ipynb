Load the dataset into a pandas dataframe. Name the variable as “survey”.

```python
import numpy as np
import pandas as pd
```

```python
survey = pd.read_excel("Sample Survey.xlsx", sheet_name='Data', index_col = 0)
survey.info()
```

How many samples were collected on each day?

```python
survey.pivot_table(index = 'collection_date', aggfunc = 'size')
```

What proportion of the total respondents were aged less than 45?

```python
survey['age'] = pd.to_numeric(survey['age'])
```

```python
age = survey['age'].to_numpy()
count = (age >45).sum()
print("Count of repondants less than 45 is:", count )
```

Create a new column in the dataframe “age_group”. This column should contain the age
group the respondent belongs to. The age groups are 18-25, 25-40, 40-55 and 55+. The
dataframe should look like this after the column creation:

```python
bins = [17, 25, 40, 55, np.inf]
lab = ['18-25','25-40','40-55', '55+']
survey['age_group'] = pd.cut(survey['age'], bins=bins, labels=lab)
```

```python
new_df = survey[['age','age_group']].copy()
new_df.head()
```

How many samples were collected for each age-group? Which age-group had the most
samples?

```python
sample_count = new_df.pivot_table(index = 'age_group', aggfunc = 'size')
sample_count
```

```python
print("Age-group having most samples: 25-40 having:", sample_count.max())
```

What proportion of the respondents had opted for the RJD party in both the Vote_Now
and the Past_Vote questions?

```python
repont = (survey[['Vote_Now', 'Past_Vote']] == "RJD").sum().min()
print("Num of Respondents opting for the RJD party in Vote_Now and Past_Vote are:", repont)
```
